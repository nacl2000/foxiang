```python
import requests
from lxml import etree
import asyncio
import os
import time
import nest_asyncio
import aiohttp
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'
}
async def download(img_src,path):
    async with aiohttp.ClientSession() as session:
        async with session.get(img_src,headers=headers) as response:
            fp=open(path,'wb')
            img=await response.read()
            #print(img)
            fp.write(img)
            fp.close()

def find_picture(num,name):
    url = 'http://diglweb.zjlib.cn:8081/zjtsg/zgsmcgzx/cypicgl.jsp?channelid=91743&searchword=%CF%F1%D6%F7%B1%E0%BA%C5='+str(num)
    params={
        'channelid': '91743'
    }
    reponse = requests.get(url=url,headers=headers,params=params)
    page_text=reponse.text
    tree=etree.HTML(page_text)
    tr_list=tree.xpath('//img[@width="75"]/@src')
    #print(len(tr_list))
    tmp=1
    tasks=[]
    for img_src in tr_list:
        img_path='./佛像/'+name+'/'+str(tmp)+'.jpg'
        #print(img_src)
        tmp+=1
        c=download(img_src,img_path)
        task =asyncio.ensure_future(c)
        tasks.append(task)
    loop = asyncio.get_event_loop()
    result=loop.run_until_complete(asyncio.wait(tasks))

start=time.time()
nest_asyncio.apply()
for i in range(566,1331):
    url = 'http://diglweb.zjlib.cn:8081/zjtsg/zgsmcgzx/cyxl.jsp?channelid=91740&record='+str(i)
    params={
        'channelid': '91740',
        'record': str(i)
    }
    reponse = requests.get(url=url, headers=headers,params=params)
    page_text=reponse.text
    tree=etree.HTML(page_text)
    title=tree.xpath('//div//tr[2]/td[3]/text()')[0]
    num=tree.xpath('//div//tr[10]/td[3]/text()')[0]
    #print(title)
    f_path='./佛像/'+str(title)
    #print(f_path)
    floder=os.path.exists(f_path)
    if not floder:
        os.mkdir(f_path)
    else:
        continue
    #print(num)
    find_picture(num,title)
end=time.time()
print(end-start)
```
